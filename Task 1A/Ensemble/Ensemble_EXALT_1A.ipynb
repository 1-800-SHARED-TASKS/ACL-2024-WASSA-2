```python
df_dev_english = df_dev_all[df_dev_all['Lang']=='English']
df_dev_spanish = df_dev_all[df_dev_all['Lang']=='Spanish']
df_dev_russian = df_dev_all[df_dev_all['Lang']=='Russian']
df_dev_french  = df_dev_all[df_dev_all['Lang']=='French']
df_dev_dutch   = df_dev_all[df_dev_all['Lang']=='Dutch']
```

```python
import pandas as pd
from sklearn.metrics import f1_score
df_dev_metrics_f1 = pd.DataFrame(columns=['Language', 'Criteria', 'gpt4_dev_attempt1', 'gpt4_dev_attempt2', 'gemma_dev_attempt1','gemma_dev_attempt2', 'claude_dev_attempt1','claude_dev_attempt2', 'mistral_dev_attempt1','mistral_dev_attempt2', 'llama_dev_attempt1','llama_dev_attempt2'])
def calculate_f1_scores(df, language):
    models = ['gpt4_dev_attempt1', 'gpt4_dev_attempt2', 'gemma_dev_attempt1','gemma_dev_attempt2', 'claude_dev_attempt1','claude_dev_attempt2', 'mistral_dev_attempt1','mistral_dev_attempt2', 'llama_dev_attempt1','llama_dev_attempt2']
    class_names = ['Anger', 'Fear', 'Joy', 'Love', 'Neutral', 'Sadness']
    criteria_types = ['micro', 'macro', 'weighted']
    for criteria in criteria_types:
        f1_scores = {'Language': language, 'Criteria': criteria}
        for model in models:
            score = f1_score(df['Labels'], df[model], labels=class_names, average=criteria)
            f1_scores[model] = score
        df_dev_metrics_f1.loc[len(df_dev_metrics_f1)] = f1_scores
datasets = {'English': df_dev_english, 'Russian': df_dev_russian, 'Spanish': df_dev_spanish, 'French': df_dev_french, 'Dutch': df_dev_dutch}
for language, df in datasets.items():
    calculate_f1_scores(df, language)
df_dev_metrics_f1
```

```python
df_dev_metrics_f1.rename(columns={'gpt4_dev_attempt1':'gpt4_1',     'gpt4_dev_attempt2':'gpt4_2',
                                  'gemma_dev_attempt1':'gemma_1',   'gemma_dev_attempt2':'gemma_2',
                                  'llama_dev_attempt1':'llama_1',   'llama_dev_attempt2':'llama_2',
                                  'claude_dev_attempt1':'claude_1', 'claude_dev_attempt2':'claude_2',
                                  'mistral_dev_attempt1':'mistral_1',  'mistral_dev_attempt2':'mistral_2',}, inplace=True)
df_dev_metrics_f1
```

# [submission 10/10] : model selection based on majority vote of the 5 models  
## incase of no mode : v3 was used

```python
from statistics import mode, StatisticsError
def majority_vote(row):
    predictions = [row['gpt4_test_preds'], row['claude_test_preds'], row['llama_test_preds'], row['mistral_test_preds'], row['gemma_test_preds']]
    try:
        return mode(predictions)
    except StatisticsError:
        return row['Final_Label_v3']
df_test_all['Final_Label_v5'] = df_test_all.apply(majority_vote, axis=1)
df_test_all['Final_Label_v5'].value_counts()
```
