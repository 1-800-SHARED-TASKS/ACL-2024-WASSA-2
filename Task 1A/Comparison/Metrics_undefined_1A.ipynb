```python
import pandas as pd
df_dev = pd.read_csv("/content/exalt_emotion_dev.tsv",sep="\t")
df_preds = pd.read_csv("/content/Emotion.tsv",sep="\t")
```

```python
df_dev
```

```python
df_preds
```

```python
df_preds = df_preds[['Labels']]
df_preds.rename(columns={'Labels':'preds'},inplace=True)
df_dev = pd.concat([df_dev,df_preds],axis=1)
df_dev
```

```python
from sklearn.metrics import f1_score
class_names = ['Angry', 'Sadness', 'Fear', 'Love', 'Joy', 'Neutral']
f1_1 = f1_score(df_dev['Labels'], df_dev['preds'], labels=class_names, average='micro')
print(f"Micro F1 Score : {f1_1:.4f}")
f1_2 = f1_score(df_dev['Labels'], df_dev['preds'], labels=class_names, average='macro')
print(f"Macro F1 Score : {f1_2:.4f}")
f1_3 = f1_score(df_dev['Labels'], df_dev['preds'], labels=class_names, average='weighted')
print(f"Weighted F1 Score : {f1_3:.4f}")
```
