```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
def plot_confusion_matrix(true_labels, predictions, model_name, ax, classes):
    cm = confusion_matrix(true_labels, predictions, labels=classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', ax=ax, xticklabels=classes, yticklabels=classes)
    ax.set_title(f'Confusion Matrix for {model_name}')
    ax.set_xlabel('Predicted Labels')
    ax.set_ylabel('True Labels')
classes = df_dev_english['Labels'].unique()
classes.sort()  # Sort the class names if needed
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))  # Adjust the size as needed
model_names = ['gpt4_old', 'gpt4_new', 'gemma', 'claude', 'mistral', 'llama3']
axes = axes.flatten()
for i, model_name in enumerate(model_names):
    plot_confusion_matrix(df_dev_english['Labels'], df_dev_english[model_name], model_name, axes[i], classes)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import f1_score
df_dev_metrics_f1 = pd.DataFrame(columns=['Language', 'Criteria', 'gpt4_old', 'gpt4_new', 'gemma', 'claude', 'mistral', 'llama3'])
def calculate_f1_scores(df, language):
    models = ['gpt4_old', 'gpt4_new', 'gemma', 'claude', 'mistral', 'llama3']
    class_names = ['Anger', 'Fear', 'Joy', 'Love', 'Neutral', 'Sadness']
    criteria_types = ['micro', 'macro', 'weighted']
    for criteria in criteria_types:
        f1_scores = {'Language': language, 'Criteria': criteria}
        for model in models:
            score = f1_score(df['Labels'], df[model], labels=class_names, average=criteria)
            f1_scores[model] = score
        df_dev_metrics_f1.loc[len(df_dev_metrics_f1)] = f1_scores
datasets = {'English': df_dev_english, 'Russian': df_dev_russian, 'Spanish': df_dev_spanish, 'French': df_dev_french, 'Dutch': df_dev_dutch}
for language, df in datasets.items():
    calculate_f1_scores(df, language)
df_dev_metrics_f1
```
